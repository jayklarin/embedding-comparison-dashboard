```python
import os, re

# Confirm working directory
os.chdir("/Users/jayklarin/__DI/Repositories/embedding-comparison-dashboard/")
print("‚úÖ Working in:", os.getcwd())

# Define source and target files
source_file = "wiki_giga_2024_100_MFT20_vectors_seed_2024_alpha_0.75_eta_0.05.050_combined.txt"
target_file = "glove_top10k_100d.txt"

# Verify that the source exists
assert os.path.exists(source_file), f"‚ùå Source file not found: {source_file}"
```

```python
# Match only lowercase dictionary-style words (no digits, punctuation, or symbols)
pattern = re.compile(r"^[a-z]+$")
```

```python
count = 0
max_words = 10000

with open(source_file, "r", encoding="utf8") as src, open(target_file, "w", encoding="utf8") as dst:
    for line in src:
        word = line.split()[0]
        if pattern.match(word):   # keep only clean words
            dst.write(line)
            count += 1
        if count >= max_words:
            break

print(f"‚úÖ Created '{target_file}' with {count:,} clean dictionary words.")
```

Cosine similarity measures how close two words are in meaning:

\[
\text{cosine\_similarity}(A, B) = \frac{A \cdot B}{|A| \, |B|}
\]

Principal Component Analysis (PCA) reduces high-dimensional vectors to 2-D for easier visualization while preserving the main variance in meaning.

Supported vector operations include:
- Addition (`+`)
- Subtraction (`-`)
- Multiplication (`*`)
- Division (`/`)
